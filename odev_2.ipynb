```python
#SHAPIRO TEST
#h0=Dagılım Normalden gelmektedir. h1=gelmemektedir.
#her parametrenin bad kolonunun 0 degeri icin normal dagılıp dagılmadıgını 
#ve bad kolonunun 1 degerleri arasında normal dagılıp dagılmadıgına bakarız 
    
        
for key in my_dict:
    for i in grps:
            stat,p=shapiro(data[my_dict[key]][data['bad']==i])
            alpha=0.05
            if(p>alpha):
                print(my_dict[key],i,"için Dagılımı Normalden Gelmektedir.h0 accepted")
            else:
                 print(my_dict[key],i," için  Dagılımı Normalden Gelmemektedir.h0 reject")
        
```

```python
#iki datanın varyansları farklımı diye bakıcaz constant(sabit,homojen) varyans vamı diye bakıyoruz
       
#barlett normal dagılımda daha iyidir.Leven normal olmayan  dagılımda daha iyidir.            
#h0 varyanslar eşittir.
#h1=varyanslardan en az bir tanesi farklıdır 
#biz h0 hipotezini isteriz
for key in my_dict:
    for i in grps:
            stat,p=levene(data['loan'][data['bad']==0],data['loan'][data['bad']==1])
            #bad kolonu 0 olanın loanı ile bad kolonu 1 olanın loanının varyanslarının farklı olupolmadıgına bakıyoruz p degerlerini alphayla kıyasalyarak.
            alpha=0.05
            if(p>alpha):
                print(my_dict[key],i,"için varyans homojendir .h0 accepted")
            else:
                print(my_dict[key],i," için  varyans homojen degildir .h0 reject")
        
           
#DEBTİNC,MORTDUE,DELİNQ,CLAGE,DEROG,NİNQ,YOJ
```

```python
#T TEST
for key in my_dict:
    for i in grps:
            stat,p=stats.ttest_ind(data['loan'][data['bad']==0],data['loan'][data['bad']==1],equal_var=True)
            #bad kolonu 0 olanın loanı ile bad kolonu 1 olanın loanının varyanslarının farklı olupolmadıgına bakıyoruz p degerlerini alphayla kıyasalyarak.
            alpha=0.05
            if(p>alpha):
                print(my_dict[key],i,"için anlamlı fark yoktur.h0 accepted")
            else:
                print(my_dict[key],i," için anlamlı fark vardır .h0 reject")

```

```python
#OLS ,tek degişkenli regresyon gibidir bir bagımlı degişken ile bir bagımsız degişkeni acıklamaa calistigimiz durumdur bir çeşit anovadır
#summarize fonksiyonunda sadece ona verilen iki kolonumu alıyoruz (x=loan y=bad gibi )
#degişkenler arasında anlamlı fark varmı diye bakıyoruz   ordinary lee square 
       
#aplha=0.05 ise;
results=smf.ols('bad ~ (loan)',data).fit() 
print(results.summary())
#p=0.240  h1 accepted yani loan parametresi anlamlı farklılık icerdiğinden modelimize ilgili parametreyi katarız  

results=smf.ols('bad ~ mortdue',data).fit()      
print(results.summary())
#p=0.260  h1 accepted yani mortdue parametresi anlamlı farklılık icerdiğinden modelimize ilgili parametreyi katarız     

results=smf.ols('bad ~ value',data).fit()    
print(results.summary())    
#p=0.160 h1 accepted yani value parametresi anlamlı farklılık icerdiğinden modelimize ilgili parametreyi katarız                
            
results=smf.ols('bad ~ yoj',data).fit()      
print(results.summary())       
#p=0.022 h0 accepted yani yoj parametresi anlamlı farklılık icermediğinden modelimize ilgili parametreyi katmayız      

results=smf.ols('bad ~ derog',data).fit()      
print(results.summary())   
#p=0.000 h0 accepted yani derog parametresi anlamlı farklılık icermediğinden modelimize ilgili parametreyi katmayız

results=smf.ols('bad ~ delinq',data).fit()       
print(results.summary())   
#p=0.000 h0 accepted yani delinq parametresi anlamlı farklılık icermediğinden modelimize ilgili parametreyi katmayız

results=smf.ols('bad ~ clage',data).fit()       
print(results.summary())   
#p=0.000 h0 accepted yani clage parametresi anlamlı farklılık icermediğinden modelimize ilgili parametreyi katmayız

results=smf.ols('bad ~ ninq',data).fit()       
print(results.summary())   
#p=0.000 h0 accepted yani ninq parametresi anlamlı farklılık icermediğinden modelimize ilgili parametreyi katmayız 

results=smf.ols('bad ~ clno',data).fit()        
print(results.summary())   
#p=0.572 h1 accepted yani clno parametresi anlamlı farklılık icerdiğinden modelimize ilgili parametreyi katarız  
        
results=smf.ols('bad ~ debtinc',data).fit() 
print(results.summary())   
#p=0.000 h0 accepted yani debtinc parametresi anlamlı farklılık icermediğinden modelimize ilgili parametreyi katmayız
```
